```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Train Data

```python
train_df = pd.read_csv('../dataset/train.csv')
display(train_df.head(8))

# Histogram of the target variable
plt.hist(train_df['Credit_Score'], bins=5)
plt.xlabel('Credit_Score')
plt.ylabel('Frequency')
plt.title('Histogram of Credit_Score [Train]')
plt.show()
```

```python
# Info
display(train_df.info())
display(train_df.describe())
```

```python
# No of Unique values
display(train_df.nunique())
```

```python
import pandas as pd

# Assuming df is your original DataFrame
# Create an empty DataFrame to store unique values
unique_values_dict = {}

# Iterate over columns
for column in train_df.columns:
    print(column)
    # # Get unique values for the current column
    unique_values = train_df[column].unique()

    # Save in a text file
    with open(f'../dataset/unique_values_train/{column}.txt', 'w') as f:
        for item in unique_values:
            f.write("%s\n" % item)
```
