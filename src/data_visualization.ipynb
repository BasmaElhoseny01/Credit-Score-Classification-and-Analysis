```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate number of rows needed
num_cols = 3
num_rows = (len(train_df.columns) + num_cols - 1) // num_cols

# Create subplots
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, num_rows*5))

# Flatten axes array
axes = axes.flatten()

# Iterate over each column
for i, col in enumerate(train_df.columns):
    print(f'Processing column {col}')
    if col == 'ID':
        continue
    # Check if column is numeric or categorical
    if train_df[col].dtype == 'object':
        # For categorical data, create a count plot
        # Show the unique categories
    
        sns.countplot(data=train_df, x=col, ax=axes[i])
        axes[i].set_title(f'Count Plot for {col}')
        axes[i].set_xticklabels(axes[i].get_xticklabels(), rotation=45)
    else:
        # For numeric data, create a histogram
        sns.histplot(data=train_df, x=col, kde=True, ax=axes[i])
        axes[i].set_title(f'Histogram for {col}')

# Hide unused subplots
for j in range(i+1, num_rows*num_cols):
    axes[j].axis('off')

# Adjust layout
plt.tight_layout()
plt.show()
```
